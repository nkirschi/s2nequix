```python
import numpy as np
import matplotlib.pyplot as plt
from nequix.model import polynomial_cutoff
```

```python
x = np.linspace(0, 15, 100)
y = polynomial_cutoff(x, r_max=6.0, p=6.0)
z = polynomial_cutoff(x, r_max=12.0, p=6.0)
plt.figure(figsize=(5, 4))
plt.plot(x, y)
plt.plot(x, z)
plt.xlabel('nuclei distance (Å)')
plt.ylabel('adjacency strength')
plt.xlim(0, 15)
plt.ylim(-0.05, 1.05)
plt.vlines([6], ymin=-0.05, ymax=1.05, color='red', linestyle='--', linewidth=1)
plt.vlines([12], ymin=-0.05, ymax=1.05, color='red', linestyle='--', linewidth=1)
plt.tight_layout()
plt.savefig('plots/polynomial_cutoff.pdf')
plt.savefig('plots/polynomial_cutoff.png')
```
