```python
import numpy as np
import matplotlib.pyplot as plt
from nequix.model import polynomial_cutoff
```

```python
plt.figure(figsize=(5, 4))
x = np.linspace(0, 15, 100)
plots = {
    c: plt.plot(x, polynomial_cutoff(x, r_max=c, p=6.0), label=f'$r_{{\\mathrm{{cut}}}} = {c}$ Å')[0]
    for c in [6.0, 8.0, 10.0, 12.0]
}
for c, line in plots.items():
    plt.vlines(c, ymin=-0.05, ymax=1.05, color=line.get_color(), linestyle='--', linewidth=1)
plt.xlabel('nuclei distance (Å)')
plt.ylabel('adjacency weight')
plt.xlim(0, 15)
plt.ylim(-0.05, 1.05)
plt.legend()
plt.tight_layout()
plt.savefig('plots/polynomial_cutoff.pdf')
plt.savefig('plots/polynomial_cutoff.png')
```
